# 1. Reduce the size of dataset by removing duplicates

```python
### Generate the processed data with no duplicates for the official fuel efficiency data from 2015 to 2023. ###
# Windows
df15 = pd.read_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/raw data/official fuel efficiency/2015_fe.csv", low_memory=False)
df16 = pd.read_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/raw data/official fuel efficiency/2016_fe.csv", low_memory=False)
df17 = pd.read_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/raw data/official fuel efficiency/2017_fe.csv", low_memory=False)
df18 = pd.read_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/raw data/official fuel efficiency/2018_fe.csv", low_memory=False)
df19 = pd.read_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/raw data/official fuel efficiency/2019_fe.csv", low_memory=False)
df20 = pd.read_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/raw data/official fuel efficiency/2020_fe.csv", low_memory=False)
df21 = pd.read_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/raw data/official fuel efficiency/2021_fe.csv", low_memory=False)
df22 = pd.read_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/raw data/official fuel efficiency/2022_fe.csv", low_memory=False)
df23 = pd.read_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/raw data/official fuel efficiency/2023_fe.csv", low_memory=False)

for df in [df15, df16, df17, df18, df19, df20, df21, df22, df23]:
    
    df.columns = df.columns.str.lower()  # convert column names to lowercase
    df.drop(columns='id', inplace=True)  # drop the 'id' column, otherwise there will be no duplicates
    
    valid_columns = df.columns[df.notnull().any()]    # pick columns that have at least one non-null value
    valid_columns = valid_columns[valid_columns.get_loc('country'):valid_columns.get_loc('cn')+1]  # pick columns that are needed to identify duplicates: from country to commercial name ! 
    df['duplicate_count'] = df.groupby(valid_columns.tolist(), dropna=False).transform('size')  # count the number of duplicates based on the selected columns, remember to include 'dropna=False' to deal with NaN values, otherwise, there will be NaN in the 'duplicate_count' column.
    
    df.drop_duplicates(subset=valid_columns.tolist(), keep = 'first', inplace=True)   # only keep the first occurrence of each duplicate, drop the rest
    df['sales'] = df['duplicate_count'] * df['r']   # calculate the total sales for each car model, NB. data from 2015 to 2019 have multiple values of 'r'

    df.to_csv(f"C:/Users/uctpql0/OneDrive - University College London/Cars data/data/processed data/official fuel efficiency/{df['year'].iloc[0]}_fe_all_nodup.csv", index=False)  # save as csv file

### combine data from 2015 to 2023
df_combined = pd.concat([df15, df16, df17, df18, df19, df20, df21, df22, df23], ignore_index=True)
df_combined.to_csv("C:/Users/uctpql0/OneDrive - University College London/Cars data/data/processed data/official fuel efficiency/2015_2023_fe_all_nodup.csv", index=False)  # save as csv file
```

# 3. Keep only passenger cars

```python
# Generate a column 'vehicle type' whose values depend on the values of 'ct' and 'cr':
# 1) if 'cr' or 'ct' contains values in ['M1S', 'M2', 'N1', 'N1G', 'N2','N2G','N3G','N3'], then vehicle type is 'commercial car';
# 2) otherwise, the vehicle type is 'unknown'
df['vehicle type'] = np.where(df['cr'].str.contains('M1S|M2|N1|N1G|N2|N2G|N3G|N3', na=False) | df['ct'].str.contains('M1S|M2|N1|N1G|N2|N2G|N3G|N3', na=False), 'commercial car', 'unknown')   # take care of nan values
# 3) among rows whose 'vehicle type is 'unknown', change the vehicle type to 'passenger car' if its of 'cr' and 'ct' contains values in ['M1', 'M1S]
df.loc[(df['vehicle type'] == 'unknown') & (df['cr'].str.contains('M1|M1S', na=False) | df['ct'].str.contains('M1|M1S', na=False)), 'vehicle type'] = 'passenger car'   # take care of nan values
# only keep rows whose 'vehicle type' is 'passenger car'
df = df[df['vehicle type'] == 'passenger car']

# df['parent'].nunique()  # 62 unique parent
```

```python
### 1. rename some of the body type: 
# Coupe as Car;
df['body type'] = df['body type'].replace('Coupe','Car')
#Sedan: Car;
df['body type'] = df['body type'].replace('Sedan','Car')
#Hatchback: Car;
df['body type'] = df['body type'].replace('Hatchback','Car')
#Coupe suv: SUV;
df['body type'] = df['body type'].replace('Coupe SUV','SUV')
#Crossover: SUV;
df['body type'] = df['body type'].replace('Crossover','SUV')
#H hatchback: Hatchback;
df['body type'] = df['body type'].replace('H Hatchback','Car')
#Minivan: MPV;
df['body type'] = df['body type'].replace('Minivan','MPV')
#Station wagon: Wagon;
df['body type'] = df['body type'].replace('Station Wagon','Wagon')
```

```python
### 2. Rename fuel type
# rename 'ft' to 'fuel type'
df.rename(columns={'ft': 'fuel type'}, inplace=True)
df['fuel type'] = df['fuel type'].str.lower()
df['fuel type'] = df['fuel type'].str.rstrip()
df['fuel type'] = df['fuel type'].str.lstrip()

# for rows with 'cn' containing 'plug' (ignoring case), replace its value in 'fuel type' by 'phev'
df.loc[df['cn'].str.contains('plug', case=False, na=False), 'fuel type'] = 'phev'
# for rows with 'cn' containing 'hybrid' (ignoring case), replace its value in 'fuel type' by 'hybrid'
df.loc[df['cn'].str.contains('hybrid', case=False, na=False), 'fuel type'] = 'hybrid'

# rename the values in 'fuel type' column
df['fuel type'] = df['fuel type'].replace('lpg','alternative_fuel')
df['fuel type'] = df['fuel type'].replace('ng-biomethane','alternative_fuel')
df['fuel type'] = df['fuel type'].replace('ng_biomethane','alternative_fuel')
df['fuel type'] = df['fuel type'].replace('e85','alternative_fuel')
df['fuel type'] = df['fuel type'].replace('biodiesel','alternative_fuel')
df['fuel type'] = df['fuel type'].replace('petrol-electric','phev')
df['fuel type'] = df['fuel type'].replace('hydrogen','alternative_fuel')
df['fuel type'] = df['fuel type'].replace('diesel-electric','phev')
df['fuel type'] = df['fuel type'].replace('petrol-gas','alternative_fuel')
df['fuel type'] = df['fuel type'].replace('hybrid/petrol/e','hybrid')      # not enough hybrid cars!?
df['fuel type'] = df['fuel type'].replace('petrol phev', 'phev')   #
df['fuel type'] = df['fuel type'].replace('petrol/electric','phev')
df['fuel type'] = df['fuel type'].replace('diesel/electric','phev')
df['fuel type'] = df['fuel type'].replace('ng','alternative_fuel')    # natural gas
df['fuel type'] = df['fuel type'].replace('cng','alternative_fuel')    # compressed natural gas
df['fuel type'] = df['fuel type'].replace('gnl','ice')    # gasoline
df['fuel type'] = df['fuel type'].replace('diesel','ice')
df['fuel type'] = df['fuel type'].replace('petrol','ice')
df['fuel type'] = df['fuel type'].replace('electric','bev')
```

```python
### 3. rename columns names: ewltp (g/km) to wltp; enedc (g/km) to nedc; m (kg) to mass; mt to mass_wltp; ep (kw) to power; ec (cm3) to capacity; electric range (km) to range
df.rename(columns={'ewltp (g/km)': 'wltp', 'enedc (g/km)': 'nedc', 
                   'm (kg)': 'mass', 
                   'mt': 'mass_wltp', 
                   'ep (kw)': 'power', 
                   'ec (cm3)': 'capacity', 
                   'electric range (km)': 'range'}, inplace=True)
```

```python
### 4.aggregate the data based on 'parent', 'brand', 'body type', 'fuel type', calculate the associated emissions level, range (if electric) for each combination.
# Group the data
df = df.groupby(['parent', 'brand', 'body type', 'fuel type']).agg({
    'wltp' : 'mean',      # for car characteristics, take the mean of different trims
    'nedc' : 'mean',
    'power' : 'mean', 
    'capacity' : 'mean', 
    'mass' : 'mean',  
    'mass_wltp' : 'mean',
    'range' : 'mean'
}).round(2).copy()
df = df.reset_index()   # Reset the index to make the groupby columns back to columns
```
